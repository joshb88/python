## Prob 1. [20 points] Square class

### Algorithm Steps:
1. Set up Square class
   - define ```__init__``` method with private attribute ```__side``` and initialize to side
   - define ```get_side and set_side``` methods
   - define ```area, perimeter, diagonal``` methods
   - define ```__str__``` method to return the label "Square with side = " and the side 
2. Create an instance of square and test each of the methods

````python
# Code for this Problem is copied into this function that will be run the in main program
class Square:

    # Init method for squard, using an argument for side
    def __init__(self, side):
        self.__side = side
    
    # A getter method for the side
    def get_side(self):
        return self.__side

    # A setter method for the side
    def set_side(self,set_side_value):
        self.__side = set_side_value
    
    # A method for determining the area
    def area(self):
        return self.__side ** 2     # Side value squared
    
    # A method for determining the perimeter
    def perimeter(self):
        return self.__side * 4      # Side value times 4 (the total number of sides)

    # A method for determining the diagonal
    def diagonal(self):
        return self.__side * (2 ** 0.5) # Side value times the square root of 2
    
    # A string method for the class
    def __str__(self):
        return (f"Square with side = {self.__side}")
````

````python
# Run your code by calling the function 
# Always test your code to check that you get what you expect
if __name__ == '__main__':
    sq = Square(side=10)
    sq.get_side()
    sq.set_side(25)
    sq.area()
    sq.perimeter()
    sq.diagonal()
    print(sq)
````

## 2. [25 points] UML Diagram, Employee and ProductionWorker Classes
Write an Employee ProductionWorker class and test the classes.  Show your design with a UML diagram.

### Algorithm Steps:
1. Set up Employee class
   - define ```__init__``` method with private attribute ```__name and __id_number```
   - define ```set_name, get_name, set_id_number, get_id_number``` methods
   - define ```__str__``` method to return the state of the object as specified in problem 
2. Create an instance of Employee and test each of the methods

````python
# Code for this Problem is copied into this function that will be run the in main program
class Employee:

    # An init method for Employee, taking values for name and ID number
    def __init__(self, name, id_number):
        self.__name = name
        self.__id_number = id_number

    
    # A setter method for name
    def set_name(self, name):
        self.__name = name
    
    # A setter method for name
    def get_name(self):
        return self.__name

    # A setter method for ID number
    def set_id_number(self, id_number):
        self.__id_number = id_number

    # A getter method for ID number
    def get_id_number(self):
        return self.__id_number

    # A string method for the class
    def __str__(self):
        return (f"Employee:\n"
                f"  Name :{self.__name}\n"
                f"  ID   :{self.__id_number}")
````
